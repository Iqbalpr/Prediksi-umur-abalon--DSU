Import library yang dibutuhkan

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
abalon = pd.read_excel ('Regresi (Excel).xlsx')
abalon
```

Prediksi umur abalone menggunakan kondisi pengukuran fisik. Umur abalone ditentukan dengan memotong cangkangnya, menodai, dan menghitung jumlah 'ring' melalui mikroskop.

```python
# Independent variabel(x) adalah Length, Diameter, Height, Whole weight, Shucked weight, Viscera weight, Shell weight.
# Dependent variabel(y) adalah Age.
```

```python
# pertama kita harus menambahkan jumlah ring dengan 1,5 untuk menunjukan umur abalon dan mengganti nama kolom menjadi Age (Umur)
abalon['Age'] = abalon['Rings']+1.5
# kedua menghilangkan kolom rings
# ketiga menghilangkan kolom sex karena ynag dibutuhkan hanya kondisi fisik
abalon2 = abalon.drop(['Rings', 'Sex'], axis = 1)
abalon2
```

```python
# keempat, melihat statistical description dari data mulai dari mean, kuartil, standard deviation dll.
abalon2.describe()
```

```python
# kelima, mencari missing values.
abalon2.isnull().sum()
#ternyata datanya sudah tidak ada missing values
```

```python
#Mengetahui nilai korelasi dari independent variable dan dependent variable.
abalon2.corr().style.background_gradient().set_precision(2)
```

```python
#keenam, buat variabel x dan y.
x = abalon2.drop(columns='Age')
y = abalon2['Age']
#Ketujuh, ucup split data menjadi training and testing dengan porsi 80:20.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=6)
#Kedelapan, bikin object linear regresi.
lin_reg = LinearRegression()
#Kesembilan, train the model menggunakan training data yang sudah displit.
lin_reg.fit(x_train, y_train)
#Kesepuluh, cari tau nilai slope/koefisien (m) dan intercept (b).
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
#Kesebelas, cari tahu accuracy score dari model menggunakan testing data yang sudah displit.
lin_reg.score(x_test, y_test)
```

```python
#Prediksi umur abalon (Length, Diameter, Height, Whole weight, Shucked weight, Viscera weight, Shell weight)
lin_reg.predict([[400,300,100,500,2000,100,20]])
```
